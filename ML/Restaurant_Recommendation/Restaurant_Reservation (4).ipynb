```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec
import lightgbm as lgb
import networkx as nx
import torch
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
import torch_geometric.transforms as T
from torch_geometric.data import Data
```

## **TF IDF**

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['cuisines'] + " " + df['highlights'])
cosine_sim = cosine_similarity(tfidf_matrix)

def recommend_restaurants_tfidf(restaurant_name, top_n=5):
    idx = df[df['name'] == restaurant_name].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    restaurant_indices = [i[0] for i in sim_scores]
    return df.iloc[restaurant_indices][['name', 'cuisines', 'aggregate_rating']]
```

```python
# Example Usage
print("TF-IDF Recommendations:")
print(recommend_restaurants_tfidf("The Flying Saucer Cafe"))
```

## 2. Word2Vec Embeddings

```python
texts = [row.split() for row in (df['cuisines'] + " " + df['highlights']).values]
word2vec_model = Word2Vec(sentences=texts, vector_size=100, window=5, min_count=1, workers=4)

def get_word2vec_vector(text):
    words = text.split()
    vectors = [word2vec_model.wv[word] for word in words if word in word2vec_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)

df['embedding'] = df['cuisines'].apply(get_word2vec_vector)
```

### 3. LightGBM Recommendation Model

```python
features = ['average_cost_for_one', 'aggregate_rating', 'votes', 'scope']
X = df[features]
y = df['aggregate_rating']  # Predicting high-rated restaurants
train_data = lgb.Dataset(X, label=y)
params = {'objective': 'regression', 'metric': 'rmse'}
model = lgb.train(params, train_data, num_boost_round=100)

def recommend_lgbm(user_budget, min_votes=10):
    filtered_df = df[df['votes'] >= min_votes]
    predictions = model.predict(filtered_df[features])
    filtered_df['predicted_score'] = predictions
    return filtered_df.sort_values('predicted_score', ascending=False)[['name', 'aggregate_rating']].head(5)
```

```python
# Example Usage
print("LightGBM Recommendations:")
print(recommend_lgbm(500))
```

```python
G = nx.Graph()
for _, row in df.iterrows():
    G.add_node(row['name'], cuisine=row['cuisines'], rating=row['aggregate_rating'])
    for other in df.itertuples():
        if row['name'] != other.name and row['cuisines'] == other.cuisines:
            G.add_edge(row['name'], other.name)

def graph_to_torch(G):
    node_mapping = {name: i for i, name in enumerate(G.nodes)}
    edge_index = torch.tensor([[node_mapping[u], node_mapping[v]] for u, v in G.edges], dtype=torch.long).t()
    x = torch.tensor([[G.nodes[n]['rating']] for n in G.nodes], dtype=torch.float)
    return Data(x=x, edge_index=edge_index)

data = graph_to_torch(G)

class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(1, 16)
        self.conv2 = GCNConv(16, 1)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

model = GCN()

def recommend_gnn(restaurant_name):
    idx = list(G.nodes).index(restaurant_name)
    output = model(data.x, data.edge_index)
    scores = output.detach().numpy().flatten()
    top_indices = scores.argsort()[-5:][::-1]
    return [list(G.nodes)[i] for i in top_indices]

# Example Usage
print("GNN Recommendations:")
print(recommend_gnn("Barbeque Nation"))
```

```python
### Graph Visualization ###
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, seed=42, k=0.3)
nx.draw(G, pos, with_labels=False, node_size=30, edge_color="gray")
labels = {n: n for i, n in enumerate(G.nodes) if i < 20}
for key, value in labels.items():
    x, y = pos[key]
    plt.text(x, y, value, fontsize=8, ha='right', va='bottom', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
plt.title("Restaurant Similarity Graph")
plt.show()
```
